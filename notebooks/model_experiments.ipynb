```python

df = pd.read_csv("../data/train/train.csv")
```

```python
df["clean_text"] = df["text"].apply(remove_punctuation)
```

```python
df["clean_text"] = df["clean_text"].apply(lambda x: [word for word in x if word.lower() not in stop_words])
```

```python
df["clean_text"] = df["clean_text"].apply(lemmatize_text)
df["clean_text"]
```

```python
def untokenize(tokens):
    return ' '.join(tokens)
df["dev_text"] = df["clean_text"].apply(untokenize)
```

```python
text = df["dev_text"].tolist()
```

```python
labels = df["author"].tolist()
```

```python
labels
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import numpy as np
```

```python
vectorizer = TfidfVectorizer(max_df=0.5, min_df=(10))
x_train = vectorizer.fit_transform(text)
```

```python
x_train_dense = x_train.toarray()
```

```python
x_train.shape
```

```python
features = vectorizer.get_feature_names_out()
```

```python
import matplotlib.pyplot as plt
```

```python
from wordcloud import WordCloud

def create_wordcloud(text, stop_words):
    if stop_words != False:
        wordcloud = WordCloud(stopwords=stop_words).generate(text)
    else:
        wordcloud = WordCloud().generate(text)

    # Display wordcloud
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")

text = df["text"].to_string()
create_wordcloud(text=text, stop_words=False)
```

```python
print(features)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train_dense, labels)
```

```python
dfp = pd.read_csv("../data/test/test.csv")
```

```python
dfp["clean_text"] = dfp["text"].apply(remove_punctuation)
dfp["clean_text"] = dfp["clean_text"].apply(lambda x: [word for word in x if word.lower() not in stop_words])
dfp["clean_text"] = dfp["clean_text"].apply(lemmatize_text)
```

```python
def untokenize(tokens):
    return ' '.join(tokens)
dfp["dev_text"] = dfp["clean_text"].apply(untokenize)
```

```python
test_text = dfp["dev_text"].tolist()
```

```python
x_test = vectorizer.transform(test_text)
```

```python
x_test = x_test.toarray()
```

```python
predictions = model.predict_proba(x_test)
```

```python
predictions
```

```python
model.classes_
```

```python
classes = model.classes_.tolist()
```

```python
classes[0]
```

```python
submission = pd.DataFrame({"id" : dfp["id"], classes[0] : predictions[:,0], classes[1] : predictions[:, 1], classes[2] : predictions[:, 2]})
```
